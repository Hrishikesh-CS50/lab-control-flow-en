## Words

```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story',  'island']
```

**Print every word in upper case**

```python
# your code here
for word in words:
    print(word.upper())
```

**Create a new list containing only words with 5 or more letters**

```python
# your code here
# create a empty list
greater_word = []

#iterate through the list to check for letters greater than 5 letters
for word in words:
 if len(word) >=5:
  greater_word.append(word) #add the letter to the greater_word 

print(greater_word)
```

**Print the first word starting with "t"**

```python
# your code here

for word in words:
    if word[0] == 't':
        print(word)
        break
```

## Numbers

**Create a list containing the square of every number from 1 to 10**

```python
# your code here
square = []

for i in range(1,11):
    square.append(i*i)

print(square)
```

**Print a list containing the square of every odd number from 1 to 10**

```python
# your code here
square = []

for i in range(1,11):
    if i%2 !=0:
        square.append(i*i)

print(square)
```

**Create a list with the squares of all multiples of 8 below 1000**

```python
# your code here
square = []
i=0

while i*i < 1000:
    if i%8 == 0:
        square.append(i*i)
    i+=1

print(square)
```
